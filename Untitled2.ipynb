```python
import pandas as pd 
```

```python
df= pd.read_csv("thebiglist_clean_extende_csv")
df
```

```python
#value counts of brand
top_thirty_brands=df['Brand'].value_counts()[:30]
top_thirty_brands
```

```python
new_brand= []

for brand in df['Brand']:
    if brand in top_thirty_brands:
        new_brand.append(brand)
    else:
        new_brand.append("other")
        
df['newBrand']= new_brand
df
        
```

```python
len(df['newBrand'].value_counts())
```

```python
#examining style
top_four_style=df['Style'].value_counts().index[:4]
#loop through style and create newstyle
new_style= []


for style in df ['Style']:
    if style in top_four_style:
        new_style.append(style)
    else:
        new_style.append("other")
df['newStyle']= new_style
df
        
```

```python
df['newStyle'].value_counts()
```

```python
len(df['Country'].value_counts())
```

```python
df['Country'].value_counts()
```

```python
len(df) * 0.8
```

```python
top_ten_countries=df['Country'].value_counts()[:10].index
top_ten_countries
```

```python
new_country= []

for country in df['Country']:
    if country in top_ten_countries:
        new_country.append(country)
    else:
        new_country.append("other")
        
df['newCountry']= new_country
df
    
```

```python
df['newCountry'].value_counts()
```

```python
#dummfying
brand_dummy= pd.get_dummies(df['newBrand'], drop_first=True ,
                           prefix= "from")
brand_dummy
```

```python
#dummfying
style_dummy= pd.get_dummies(df['newStyle'], drop_first=True ,
                           prefix= "is")
style_dummy
```

```python
#dummfying
country_dummy= pd.get_dummies(df['newCountry'], drop_first=True ,
                           prefix= "in")
country_dummy
```

```python
df_temp=df.drop(['Brand','Variety','Style','Country'
                 ,'newBrand','newStyle','newCountry'], axis=1)
df_temp
```

```python
df_final= pd.concat([df_temp, brand_dummy,style_dummy, country_dummy],
                    axis=1)
df_final
```

```python
df_final.to_csv('thebiglist_clean_extended_final.csv', index=None)
```
